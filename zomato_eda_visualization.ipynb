Zomato Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe=pd.read_csv("Zomato_dataset.csv")
dataframe
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0]
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
dataframe.head()
```

```python
dataframe.info()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('Type of Restaurant')
```

```python
#conclusion: max order from dining type of restaurant
```

```python
group=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':group}) #created new dataframe contains only two columns
plt.plot(result, c='green',marker='o')
plt.xlabel('Type of Restaurant')
plt.ylabel('Votes')
```

```python
#conclusion: dining restaurant has received max votes
```

```python
plt.hist(dataframe['rate'],bins=7)
plt.title('Rating Distribution')
plt.show()
```

```python
#conclusion: majority of rating received between 3.25 to 4.0
```

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['approx_cost(for two people)'])
```

```python
#conclusion: majority of couples prefer to spend 300 on an order
```

```python
sns.boxplot(x='online_order',y='rate',data=dataframe)
```

```python
#conclusion: online orders have higher rating than offline
```

```python
pivot_table= dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu',fmt='d')
plt.title('Heatmap')
plt.xlabel('Online order')
plt.ylabel('Listed in(type)')
plt.show()
```

```python
#conclusion: people prefer offline at dining restaurant, order online at cafes
```
